# 1. Carregamento dos dados

Nesta etapa, importamos as bibliotecas necessárias para análise de dados e visualização. Também é adidcionado o dataset Titanic e estrtutura geral

As principais bibliotecas utilizadas são:
- pandas: manipulação de dados
- matplotlib e seaborn: visualização gráfica

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("train.csv")
df.head()
df.shape
df.columns
```

O dataset possui informações como:

- Survived: indica se o passageiro sobreviveu (0 = não, 1 = sim)
- Pclass: classe do ticket (1ª, 2ª, 3ª)
- Sex: sexo
- Age: idade
- SibSp: número de irmãos/cônjuges a bordo
- Parch: número de pais/filhos a bordo
- Fare: tarifa paga
- Embarked: porto de embarque

Aqui vemos o tipo das variáveis.

- Numéricas: valores quantitativos (ex: idade, tarifa) = Age, Fare, SibSp, Parch
- Categóricas: representam grupos (ex: sexo, classe) = Sex, Embarked, Pclass

```python
df.dtypes
df.describe()
```

**2. Analisando a variável "Survived"**

```python

# Frequência
df['Survived'].value_counts()
# Tabela de frequência
df['Survived'].value_counts(normalize=True)
# Gráfico
sns.countplot(x='Survived', data=df)
plt.title("Distribuição de Sobreviventes")
plt.xlabel("Sobreviveu (0 = Não, 1 = Sim)")
plt.ylabel("Quantidade")
plt.show()
```

Resposta:
- Aproximadamente 342 pessoas sobreviveram
- A classe mais frequente é a de não sobreviventes (0)

**Relação entre "Sex" e "Survived**

```python
# Tabela cruzada
pd.crosstab(df['Sex'], df['Survived'])
# Frequência relativa
pd.crosstab(df['Sex'], df['Survived'], normalize='index')
# Gráfico
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Sobrevivência por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Quantidade")
plt.show()
```

As mulheres sobreviveram significativamente mais que homens.
o que possivelmente indica um critério de prioridade no resgate.

Distribuição das **idades**

```python
# Verifica os valores nulos
df['Age'].isnull().sum()
# faz a mediana
df['Age'] = df['Age'].fillna(df['Age'].median())
# Histograma
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Distribuição das Idades")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()
```

A maioria dos passageiros era jovem, com maior densidade entre 20 e 40 anos.
